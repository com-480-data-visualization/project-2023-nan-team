```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
data  = pd.read_csv("data/swiss_all.csv")
print('nbr of datapoints per year',len(data[data['year']==2000]))
```

```python
mean_density = []
for i in range(2000,2021):
    mean_density.append(np.mean(data[data['year']==i]["Z"]))
plt.plot(list(range(2000,2021)),mean_density,label='Swiss')
plt.title("Mean density through the year")
plt.xlabel("year")
plt.ylabel("density pop/km^2")
plt.legend()
plt.xticks(np.linspace(2000,2020,5))
plt.show()
```

```python
print("max density: ",data.iloc[np.argmax(data['Z'])]['Z'])
print("min density: ",data.iloc[np.argmin(data['Z'])]['Z'])
print("max latitude: ",data.iloc[np.argmax(data['X'])]['X'])
print("min latitude: ",data.iloc[np.argmin(data['X'])]['X'])
print("max longitude: ",data.iloc[np.argmax(data['Y'])]['Y'])
print("min longitude: ",data.iloc[np.argmin(data['Y'])]['Y'])
```
